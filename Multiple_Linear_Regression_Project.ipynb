```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# loading dataset
df= pd.read_csv('https://raw.githubusercontent.com/Priyanka797/car-price-prediction-linear-regression/refs/heads/main/CarPrice_Assignment.csv')
```

```python
# Explore dataset
df.head()
df.tail()
df.info()
df.describe()
df.shape
df.size
df.max
df.min
```

```python
# Handling missing values
df.isnull().sum()
```

```python
#Handling duplicate value
print(f"The duplicate: {df.duplicated().sum()}")
```

```python
# dropping th User ID column
data_new = df.drop(['CarName','carbody', 'enginetype','cylindernumber', 'fuelsystem','drivewheel' ] , axis=1)
data_new.head()
```

```python
#one hot encoding
categorical_cols = ['fueltype', 'aspiration', 'doornumber','enginelocation']

data_new_encoded = pd.get_dummies(data_new, columns=categorical_cols, drop_first=True)
```

```python
X = data_new[['car_ID', 'symboling', 'wheelbase', 'carlength',
        'carwidth', 'carheight', 'curbweight',
        'enginesize', 'boreratio', 'stroke', 'compressionratio',
        'horsepower', 'peakrpm', 'citympg', 'highwaympg']]

y = data_new['price']
```

```python
#initialize model
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
print(y_pred)
print(y_test)
```

```python
print(model.coef_)
print(model.intercept_)
```
